```python
#define:Delete all retweets and the records without images
#code:
twitter_archive_master = twitter_archive_master[pd.isnull(twitter_archive_master.retweeted_status_id)]
twitter_archive_master = twitter_archive_master.dropna(subset = ['jpg_url'])
```

```python
#define: Delete the columns that we don't need anymore
#code:
twitter_archive_master = twitter_archive_master.drop('retweeted_status_id', 1)
twitter_archive_master = twitter_archive_master.drop('retweeted_status_user_id', 1)
twitter_archive_master = twitter_archive_master.drop('retweeted_status_timestamp', 1)
twitter_archive_master = twitter_archive_master.drop('date_time', 1) #its repeated!
```

```python
#define: delete the 4 columns of dog stages and make it 1 instead
#code:
melt = ['doggo','floofer','pupper','puppo']
rest = [x for x in twitter_archive_master.columns.tolist() if x not in melt]
twitter_archive_master = pd.melt(twitter_archive_master, id_vars = rest , value_vars = melt, var_name = 'stages', value_name = 'dog_stage')
twitter_archive_master = twitter_archive_master.drop('stages', 1)
twitter_archive_master = twitter_archive_master.sort_values('dog_stage').drop_duplicates('tweet_id', keep = 'last')
```

```python
#test
twitter_archive_master.info()
```

```python
#define: edit the invalid names of the dogs
#code:
import re
names = []

for t in twitter_archive_master ['text']:
    if t.startswith('This is ') and re.match(r'[A-Z].*', t.split()[2]):
        names.append(t.split()[2].strip(',').strip('.'))
    elif t.startswith('Meet ') and re.match(r'[A-Z].*', t.split()[1]):
        names.append(t.split()[1].strip(',').strip('.'))
    elif t.startswith('Say hello to ') and re.match(r'[A-Z].*', t.split()[3]):
        names.append(t.split()[3].strip(',').strip('.'))
    elif t.startswith('Here we have ') and re.match(r'[A-Z].*', t.split()[3]):
        names.append(t.split()[3].strip(',').strip('.'))
    elif 'named' in t and re.match(r'[A-Z].*', t.split()[t.split().index('named') + 1]):
        names.append(t.split()[t.split().index('named') + 1].strip(',').strip('.'))
    else:
        names.append('NaN')
twitter_archive_master['names'] = names
```

```python
#delete the old column of the name
twitter_archive_master= twitter_archive_master.drop(['name'], 1)
```

```python
#test:
twitter_archive_master.info()
```

```python
#define: edit the invalid rate of the dogs by taking all the exact rate from the text ot tweet
#code
rating = twitter_archive_master['text'].apply(lambda i: re.findall(r'(\d+(\.\d+)|(\d+))\/(\d+0)', i))
rating_num = []
rating_den = []

for rate in rating:
    if len(rate) == 0:
        rating_num.append('NaN')
        rating_den.append('NaN')
    
    elif len(rate) == 1:
        rating_num.append((float(rate[0][0]) / (float(rate[0][-1])/10)))
        rating_den.append(float(rate[0][-1]))
                          

    elif len(rate) > 1 and rate[0][-1] == '10':
        plus = 0
        avgerage = 0
        for x in range(len(rate)):
            plus = plus + float(rate[x][0])
        result = (plus / len(rate))

        rating_num.append(result)
        rating_den.append(10)

    else: 
        rating_num.append('error')
        rating_den.append('error')
```

```python
#add the privious rates into our data
twitter_archive_master['rating_num'] = rating_num
twitter_archive_master['rating_den'] = rating_den
```
