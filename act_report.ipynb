**Visualization**

After we finish all the wrangling steps of the data then store it as csv file, now we import the cleaning data and start the visualization step.<br>
<br>
First, We draw a scatter plot that represent the relation between the two variables - retweets number and favorites number - to see if there is some relation between them, and weather its positive or not.<br>
<br>

Second, We represent the stages of dogs in pie chart to see the percentage that each stage is represent in our dataset.<br>
<br>

Finally, We calculate the rating mean for each group of dog stage to see if the rating have a big change according to stage or not, then look if there is a positive relation between the majority of dog stage and their rating.<br>
<br>
As we can notice from the previous visualization of the data:<br>
1- There is a positive relation between the number of retweets and number favorites of the tweets. Which mean that as long as the retweets number increase, favorites number will increase also and vise versa.<br>
2- The vast majority of the rated dogs are Pupper, which represent 69.3 - almost three quarters of all stages -.<br>
3- There are not a huge difference between the rating of each stage, especially between doggo and floofer.<br>
4- The rating mean of each stage does not have any relation with number of dogs within the stage, so although that vast majority is for Pupper , the Puppo have higher rating mean than Pupper and the other  stages.<br>
